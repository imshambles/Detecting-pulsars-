```python
df = pd.read_csv("Train.csv")
df.info()
```

```python
import seaborn as sb
```

```python
#plot mean_profile VS Pulsar
sb.lmplot(x = 'Mean_Profile', y = 'Pulsar', data = df)
```

```python
#Plot mean_curve VS pulsar
sb.lmplot(x = 'Mean_Curve', y = 'Pulsar', data = df)
```

```python
#plot sd_profile VS pulsar
sb.lmplot(x = 'SD_Profile', y = 'Pulsar', data = df)
```

```python
#plot sd_curve VS pulsar
sb.lmplot(x = 'SD_Curve', y = 'Pulsar', data = df)
```

```python
#plot Excess_kurtosis_Profile VS pulsar
sb.lmplot(x = 'Excess_kurtosis_Profile', y = 'Pulsar', data = df)
```

```python
#plot Excess_kurtosis_Curve VS pulsar
sb.lmplot(x = 'Excess_kurtosis_curve', y = 'Pulsar', data = df)
```
